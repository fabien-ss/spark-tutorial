# Compréhension de Spark Caching


```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from datetime import datetime
```

Création de l'instance spark, getOrCreate créée ou utilise une instance de spark déjà créée

```python
spark = SparkSession.builder.appName("ColabSpark").getOrCreate()
```

Création d'un jeu de données

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import rand, monotonically_increasing_id
import pyspark.sql.functions as F

# Créer un DataFrame avec 1 000 000 de lignes
data_test = spark.range(10000000) \
    .withColumn("valeur_float", rand()) \
    .withColumn("date_dispo", F.lit("20250219")) \
    .withColumn("date_dispo1", F.lit("20250218")) \
    .withColumn("date_dispo2", F.lit("20250119")) \
    .withColumn("date_dispo3", F.lit("20250119")) \
    .withColumn("date_dispo4", F.lit("20250119")) \
    .withColumn("date_dispo5", F.lit("20250119")) \
    .withColumn("date_dispo6", F.lit("20250119")) \
    .withColumn("date_dispo7", F.lit("20250119")) \
    .withColumn("date_dispo8", F.lit("20250119")) \
    .withColumn("date_dispo9", F.lit("20250119")) \
    .withColumn("date_dispo10", F.lit("20250119")) \

# Sauvegarder en Parquet
data_test.write.mode("overwrite").parquet("/content/drive/MyDrive/Spark/date_test.parquet")
data_test.count()
```

```python
df = spark.read.parquet("/content/drive/MyDrive/Spark/date_test.parquet")

df.cache()

# df = df.filter(F.col("date_dispo") == "20250219")

df.count()
```

```python
test_1_start_date = datetime.now()

df.show(n=5)

test_1_end_date = datetime.now()

print(f"temps d'éxecution {test_1_end_date - test_1_start_date}")
```
