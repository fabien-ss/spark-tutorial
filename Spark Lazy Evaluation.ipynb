## How Lazy Evaluation Works

**Transformations (Lazy):** Operations like filter(), select(), withColumn(), groupBy(), and orderBy() are considered transformations. When you apply a transformation, Spark simply records the operation in the DAG (lineage graph) but does not process any data.

**Actions (Eager):** Operations like collect(), count(), show(), take(), and save() are actions. When an action is invoked, it forces Spark to execute the entire accumulated DAG and return the results to the driver program or an external storage system.

```python
df = df.filter(F.col("_c4").contains("a"))
df = df.filter(F.col("_c5").contains("um"))
```
